To produce a puntuation, for each user this cose:
- Creates a song punctuation depending on the times the song has been reproduced
- For every artist, find on spotipy the genres that have assigned
- If the artist, have more than one genre, just the most listened by the user is keeped
- Creates another song punctuation depending on the popularity of the genre of its artists
- Merges the two puncutations by a ponderation

(It usually takes around 18 minutes to compute)

```python
list_punt=[]
for j in range(num_users):
  u=j
  df_hist=df_hist_total.loc[df_hist_total['User']==(u)]

#Ajuntem les cançons per grup:
  df_grups=df_hist.groupby(by=['artistName']).sum()
  df_grups=df_grups.sort_values(by=['msPlayed'], ascending = False)
  df_grups = df_grups.reset_index(level=0)
  df_grups

#busquem els generes de cada artista:
  list_grups=df_grups.values.tolist()
  list_grups=df_grups['artistName'].values.tolist()

  new_list=[]
  sht=[]

  for i in range(0, len(list_grups)):
    artista=list_grups[i]
    genres_i= sp.search(q=artista, type='artist', limit=50)
    if genres_i['artists']['items']==[]:
      new_list.append('non')
    elif genres_i['artists']['items'][0]['genres']==[]:
      sht.append(artista)
      new_list.append('non')
    else:
      miau=genres_i['artists']['items'][0]['genres']
      new_list.append(miau)
  #new_list
  df_grups['Genres']=new_list
  df_grups.head()

  ####  PUNTUACIÓ PER COPS ESCOLTATS ####

  #Ajuntem les cançons per grup:
  df_songs=df_hist.groupby(by=['trackName'], as_index=False).size()
  df_songs=df_songs.sort_values(by=['size'], ascending = False)
  list_songs=df_songs['trackName'].values.tolist()
  s_1=df_songs.loc[df_songs['trackName']==list_songs[0]]
  times_max=s_1['size'].tolist()[0]
  max=len(df_songs) 
  step=times_max/7    
  time_p={}
  for i in range(len(list_songs)):
    times_song=df_songs.loc[df_songs['trackName']==list_songs[i]]['size'].values
    if i <=max/5:
      puntuation = 5
    elif i<=2*max/5:
      puntuation = 4
    elif i<=3*max/5:
      punctuation = 3
    elif i<= 4*max/5:
      punctuation = 2
    else:
      puntuation = 1
    time_p[list_songs[i]]=puntuation

    #### PUNTUACIO PELS COPS QUE EL GÈNERE HA ESTAT ESCOLTAT #####

  #com que treballem amb el df_grups que surten els grups un cop, cal multiplicar els cops que es repeteix un genere per el nombre de cops que s'ha repetit l'artista que te el genere:
  list_genres=df_grups['Genres'].values.tolist() #genres de la llista dels artistes
  list_artists=df_grups['artistName'].values.tolist()
  #times the artist - so its genres - have appeared:
  df_artist=df_hist['artistName'].value_counts()
  dicc_genres={}
  l_g=[]
  for i in range(len(list_genres)):
    for j in range(len(list_genres[i])):
      a=list_artists[i]
      g=[list_genres[i][j]][0]
      a_times=df_artist[a] #es el que li sumem, 
      ll=['artist:', a, 'appeared times: ', a_times, 'genre', g]
      l_g.append(ll)
      if list_genres[i][j] in dicc_genres:
        dicc_genres[list_genres[i][j]]=dicc_genres[list_genres[i][j]]+a_times
      else:
        dicc_genres[list_genres[i][j]]=a_times

  #de cada artista, puntua del 1 al 5 el seu genere que ha aparegut mes vegades:      
  #de cada genere hi assignem una puntuació:
  #list_artists i list_genres
  punt_genres=[]
  dd_genre={}
  dd_punt=[]
  for i in range(len(list_artists)):
    #locate the artist on the df_grups and extract the list of genres g for the i'th artist
    a=df_grups.loc[df_grups['artistName']==list_artists[i]]
    g=a['Genres'].values.tolist()[0]

    #cada llista de generes pot tenir varis elements, ens quedem amb la millor puntuacio:
    p=0

    #Revisem el genere millor que ha sigut mes escoltat de cada genere
    for j in range(len(g)):
      p_0=dicc_genres[g[j]]
      best_g=g[j]
      if best_g == 'non':
        p=0;
      elif p_0 > p:
        p = p_0;
        best_g=g[j]    
    dd_genre[list_artists[i]]= best_g
    dd_punt.append([list_artists[i], dicc_genres[best_g]])

  #Ara tenim una llista on surt cada artista amb el seu genere mes ben puntuat - dd
  sorted_punt = sorted(dd_punt, key=lambda x: x[1], reverse = True)
  max=len(sorted_punt)
  genre_p={}
  for i in range(len(sorted_punt)):
    if i < max/5:
      punt=5
    elif i < 2*max/5:
      punt = 4
    elif i < 3*max/5:
      punt = 3
    elif i < 4*max/5:
      punt = 2
    else:
      punt = 1
    genre_p[sorted_punt[i][0]]=punt




  #ponderem les dues ponderacions: time_p (song) //// genre_p (group)
  percentatge_times = 0.6
  percentatge_genre=0.4
  
  for i in range(df_hist.shape[0]):
    art=df_hist.iloc[i]['artistName']
    song=df_hist.iloc[i]['trackName']
    g_p=genre_p[art]
    s_p=time_p[song]
    puntuacio=s_p*percentatge_times + g_p*percentatge_genre
    list_punt.append(round(puntuacio))

#finalment ajuntem tota la llista de puntuacio
df_hist_total['Punctuation']=list_punt #### eror 
```
